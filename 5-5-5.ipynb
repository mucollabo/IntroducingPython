```python
import itertools
```

```python
for item in itertools.chain([1, 2], ['a', 'b']):
    print(item)
```

```python
for item in itertools.cycle([1, 2]):
    print(item)
```

```python
for item in itertools.accumulate([1, 2, 3, 4]):
    print(item)
```

```python
def multiply(a, b):
    return a * b
```

```python
for item in itertools.accumulate([1, 2, 3, 4], multiply):
    print(item)
```
